#### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Generate Coordinates

```python
z = np.linspace(0,1,100)
x = z * np.sin( 25 * z)
y = z * np.cos(25 * z)

fig = plt.figure()
ax = plt.axes(projection = '3d')


# ax.plot3D(x,y,z, 'green')
ax.plot3D(x,y,z)
plt.show()
```
