#### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Generate Coordinates

```python
z = np.linspace(0,1,300)
x = z * np.sin( 25 * z)
y = z * np.cos(25 * z)
# z = np.linspace(0,0,len(x)) # use this if there are no non-zero z coordinates

fig = plt.figure()
ax = plt.axes(projection = '3d')

ax.plot3D(x,y,z)
plt.show()
```

#### Save  the coordinates generated above into external text file

```python
# this data is reliant on external .txt files to save coordinates too
# change paths as required

read_data_x = open("coord_data_x.txt","w")

read_data_y = open("coord_data_y.txt","w")

read_data_z = open("coord_data_z.txt","w")

for i in range(len(x)):
    y_list = read_data_y.write(str(y[i]) + ',')
    x_list = read_data_x.write(str(x[i]) + ',')
    z_list = read_data_z.write(str(z[i]) + ',')
read_data_y.close()
read_data_x.close()
read_data_z.close()
```
